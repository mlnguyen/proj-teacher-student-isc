```python
# python imports
from os import listdir, makedirs, walk
from os.path import isfile, join, exists
import subprocess
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%config InlineBackend.figure_format = 'retina' # for 4k screen
from IPython.core.interactiveshell import InteractiveShell # for var view
InteractiveShell.ast_node_interactivity = "all" # for var view
import shutil
import pandas as pd
```

```python
#define paths
const_scan_data_path = '/mnt/bucket/labs/hasson/meshulam/onlineL/pilot/scans/'
const_process_data_path = '/mnt/bucket/labs/hasson/meshulam/onlineL/pilot/process/'
```

```python
# copy and split raw dicoms to new directories + create index file (session, number of dicoms)

#subjects = ['BS_082917','JY_091917','TC_091917','BT_092617','CK_092617','RT_092817','CN_092817','LB_100317','AL_100517','EW_100917','AA_100917','NP_101017','SN_101317','AT_101617'.'SW_102017','JM_102017','RB_102017','MP_102317','EL_112917','AF_120517','HR_121117','BE_121517']
subjects = []

# go over subjects
for this_subject in subjects:
    # raw data folder
    path_raw_data = const_scan_data_path + this_subject + '/dcm/'
    # new data folder: make folder for subject
    path_proc_files = const_process_data_path+this_subject+'/dcm/'
    if not exists(path_proc_files):
        makedirs(path_proc_files)
    
    # find all files in this scan
    scan_files = [f for f in listdir(path_raw_data) if isfile(join(path_raw_data, f))]
    #display(scan_files)
    
    # get number of files in each session, save as csv file (col1 = session, col2 - number of dicoms)
    temp = [int(str(f).split('-')[0]) for f in scan_files]  
    numof_files_in_each_session = np.asarray([[x,temp.count(x)] for x in set(temp)]) 
    pd.DataFrame(numof_files_in_each_session).to_csv(path_proc_files+this_subject+'_index.csv', index=False)
        
    # make folder for each session, move files there if not copied already
    for iScan in numof_files_in_each_session[:,0]:
        proc_session_dir = path_proc_files+'s_'+str(iScan)+'/'
        if not exists(proc_session_dir):
            makedirs(proc_session_dir)      
        files_in_session = [f for f in scan_files if int(str(f).split('-')[0])==iScan] # list all files from this session
        for iFile in files_in_session:
            if not exists(proc_session_dir+iFile):
                shutil.copy2(path_raw_data+iFile, proc_session_dir);
```

```python
#extract order of video clips and show on screen H:history, C:cs, P:pieman

#read list of subjects (datasets) from file
subjects = pd.read_csv(const_process_data_path + 'pilot_datasets_to_process.txt')['subjects'].tolist()
# prepare list with order of videos
video_order_all_subjects = []

# go over all subjects
for this_subject in subjects:
    # prepare empty string for order of videos
    video_order_this_subject = ''
    # path to logfile
    path_log1_data = const_scan_data_path + this_subject + '/logs/'
    # find relevant logfile (shortest with the CPH sequence data)
    find_file = [f for f in listdir(path_log1_data) if 'questions.csv' in f]
    if len(find_file) != 1: print('logfile error')
    #read
    log_contents = pd.read_csv(path_log1_data + find_file[0])
    path_slide5 = [l for l in log_contents['questionPath'] if 'Slide5' in l] # one per loop #slide5: could be any slide that appears in all three
    # update this subject order
    for this_file in path_slide5:
        if 'pieman' in this_file:
            video_order_this_subject += 'P'
        elif 'cos' in this_file:
            video_order_this_subject += 'C'
        elif 'hist' in this_file:
            video_order_this_subject +='H'
    #update all subjects order
    video_order_all_subjects.append(video_order_this_subject)

# show order on screen
print(subjects)
print(video_order_all_subjects)
    
```
